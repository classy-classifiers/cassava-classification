```python
import tensorflow as tf
from tensorflow import keras
```

```python
RESNET_50_PATH = '/content/drive/Shareddrives/2040_Midterm_Project/Classy_Classifiers/test_checkpoints/ResNet_final.hdf5'
INCEPTION_PATH = '/content/drive/Shareddrives/2040_Midterm_Project/Classy_Classifiers/test_checkpoints/InceptionNet_final.hdf5'
EFFICIENT_PATH = '/content/drive/Shareddrives/2040_Midterm_Project/Classy_Classifiers/test_checkpoints/EfficientNet_final.hdf5'
```

```python
resnet_50_model = tf.keras.models.load_model(RESNET_50_PATH)
efficient_model = tf.keras.models.load_model(EFFICIENT_PATH)
inception_model = tf.keras.models.load_model(INCEPTION_PATH)

models = [resnet_50_model.output, efficient_model.output, inception_model.output]
```

***The layers in these models have the same default names -- giving them new unique names:***

```python
resnet_50_model._name = 'resnet_mod'
inception_model._name = 'inception_mod'
efficient_model._name = 'efficient_mod'
```

```python
input = tf.keras.layers.Input(shape=(360,360,3))
x1 = resnet_50_model(input)
x2 = efficient_model(input)
x3 = inception_model(input)
ensemble_layer = keras.layers.Add()([x1,x2,x3])
ensemble_model = keras.Model(inputs = [input], outputs = ensemble_layer)


# keras.utils.plot_model(ensemble_model, "ensemble_architecture.png", show_shapes=True)
```
